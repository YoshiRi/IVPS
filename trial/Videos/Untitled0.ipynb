```python
import cv2
import numpy as np
import sys
import math
import matplotlib.pyplot as plt
sys.path.append("../../")
%matplotlib inline
```

```python
from vmarker import *
```

```python
K = np.loadtxt("../../calib_usb/K.csv",delimiter=",")
dist_coef = np.loadtxt('../../calib_usb/d.csv',delimiter=",")
```

```python
cap = cv2.VideoCapture('x-05y-05.avi')
vm = vmarker(K=K,dist=dist_coef,markerpos_file="../roomA_ground_orig.csv")
rvs = []
tvs = []

while 1:
    ok,frame = cap.read()
    if not ok:
        print("Finished!")
        break
    sframe = frame.copy()
    cv2.startWindowThread()
    tv=vm.getcamerapose(frame)
    cv2.waitKey(1)
    if vm.PNPsolved:
        rvs.append(vm.rvecs)
        tvs.append(vm.tvecs)
        
```

```python
tvplot = np.array(tvs).reshape(-1,3)
rvplot = np.array(rvs).reshape(-1,3)

fig=plt.figure(figsize=(18, 16), dpi= 80, facecolor='w', edgecolor='k')
plt.subplot(231)
plt.plot(tvplot[:,0])
plt.subplot(232)
plt.plot(tvplot[:,1])
plt.subplot(233)
plt.plot(tvplot[:,2])
plt.subplot(234)
plt.plot(rvplot[:,0])
plt.subplot(235)
plt.plot(rvplot[:,1])
plt.subplot(236)
plt.plot(rvplot[:,2])
```
