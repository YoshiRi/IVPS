## Extract leftimage messages from rosbag


```python
import rosbag
import cv2
from cv_bridge import CvBridge, CvBridgeError
from sensor_msgs.msg import Image
```

```python
timestamps = [];
images = [];
for topic, msg, t in  rosbag.Bag('/home/yoshi/Downloads/firsttry0402.bag').read_messages():
    if topic == '/zed/left/image_raw_color':
        timestamps.append(t.to_sec())
        images.append( CvBridge().imgmsg_to_cv2(msg, "8UC3") )
```

```python
plt.imshow(images[8])
```

```python
ref6image = cv2.imread("imgdump/frame0006.jpg")

for i in range(len(images)):
    print(SSD(ref6image,images[i]))
    if SSD(ref6image,images[i]) < 100:
        print(i)
        break
```

```python
def SSD(i1c,i2c):
    i1 = cv2.cvtColor(i1c,cv2.COLOR_BGR2GRAY)
    i2 = cv2.cvtColor(i2c,cv2.COLOR_BGR2GRAY)
    assert i1.shape == i2.shape
    di2 = (i1-i2)*(i1-i2)
    return sum(sum(di2))


SSD(left,right)
```

```python
## Create Video for Evaluation
size = (left.shape[1],left.shape[0])
out = cv2.VideoWriter('project.avi',cv2.VideoWriter_fourcc(*'DIVX'), 15, size)
for i in range(len(images)):
    out.write(images[i])
out.release()
```
